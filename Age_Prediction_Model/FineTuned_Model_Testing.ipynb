```python
import gdown
import tensorflow as tf
from facenet_pytorch import MTCNN
from PIL import Image, ImageDraw
import numpy as np
import cv2
from matplotlib import pyplot as plt
import os
from tensorflow.keras.applications.resnet50 import preprocess_input

# -----------------------------
# 1. Load the fine-tuned ResNet50 model
# -----------------------------
file_id = "1xyGYBYRQwjBpPreZmvv_enQ3Z92wHNSx"
download_url = f"https://drive.google.com/uc?id={file_id}"
output_path = "resnet50_age_model.keras"
gdown.download(download_url, output_path, quiet=False)
model = tf.keras.models.load_model(output_path)
print("Model loaded!")

# -----------------------------
# 2. Load MTCNN for face detection
# -----------------------------
mtcnn = MTCNN()

# -----------------------------
# 3. Face alignment function
# -----------------------------
def align_face(img_cv, landmark, output_size=(224,224)):
    """
    Align face based on eyes and nose landmarks (for ResNet50 input size 224x224)
    """
    left_eye, right_eye, nose = landmark[0], landmark[1], landmark[2]

    # Compute center between eyes
    eyes_center = ((left_eye[0]+right_eye[0])/2, (left_eye[1]+right_eye[1])/2)

    # Angle between eyes
    dy, dx = right_eye[1]-left_eye[1], right_eye[0]-left_eye[0]
    angle = np.degrees(np.arctan2(dy, dx))

    # Scale
    desired_eye_distance = output_size[0] * 0.35
    dist = np.sqrt(dx**2 + dy**2)
    scale = desired_eye_distance / dist

    # Rotation matrix
    M = cv2.getRotationMatrix2D(eyes_center, angle, scale)

    # Translate to center
    tX, tY = output_size[0]*0.5, output_size[1]*0.4
    M[0,2] += (tX - eyes_center[0])
    M[1,2] += (tY - eyes_center[1])

    aligned_face = cv2.warpAffine(img_cv, M, output_size, flags=cv2.INTER_CUBIC)
    return aligned_face

# -----------------------------
# 4. Preprocessing function for ResNet50
# -----------------------------
def preprocess_for_model(face_img):
    x = cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB)  # OpenCV -> RGB
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return x

```

# Testing some samples on fine-tuned model

```python
# -----------------------------
# 5. Process a test image folder
# -----------------------------
test_folder = "test_images"

for img_name in os.listdir(test_folder):
    img_path = os.path.join(test_folder, img_name)

    if os.path.isdir(img_path):
        continue

    # Load image
    img = Image.open(img_path).convert('RGB')  # <--- Important
    img_cv = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
    # Detect faces
    boxes, probs, landmarks = mtcnn.detect(img, landmarks=True)

    if boxes is None:
        print(f"No faces detected in {img_name}")
        continue

    # For each detected face
    for i, landmark in enumerate(landmarks):
        aligned_face = align_face(img_cv, landmark, output_size=(224,224))
        x_input = preprocess_for_model(aligned_face)

        # Predict age
        pred_age = model.predict(x_input)[0][0]

        # Display face with predicted age
        plt.figure()
        plt.imshow(cv2.cvtColor(aligned_face, cv2.COLOR_BGR2RGB))
        plt.title(f"{img_name} - Face {i+1} - Predicted Age: {pred_age:.1f}")
        plt.axis('off')
        plt.show()
```
